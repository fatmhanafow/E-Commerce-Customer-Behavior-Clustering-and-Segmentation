```python
%%writefile preprocessor.py
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder

class Preprocessor:
    def __init__(self):
        self.scaler = StandardScaler()
        self.city_encoder = LabelEncoder()
        self.features_to_scale = None
        self.current_date = None
        
    def fit(self, df):
        # Learns the necessary parameters to transform the data.
        # data in order level -> customer level
        customer_df = self._create_customer_features(df)
        
        # Learning encoding for the city
        self.city_encoder.fit(customer_df['city_name_fa'])
        
        # Determine the columns that need to be scaled
        self.features_to_scale = customer_df.select_dtypes(include=[np.number]).columns.tolist()
        
        # remove the customer ID from the list.
        if 'ID_Customer' in self.features_to_scale:
            self.features_to_scale.remove('ID_Customer')

        # Learning for Standardization
        self.scaler.fit(customer_df[self.features_to_scale])

        return self

    def transform(self, df):
        # function applies the learned transformations to the data.
        # Creating Customer Level Features
        customer_df = self._create_customer_features(df)
        
        # Convert City to Number
        customer_df['city_name_fa'] = self.city_encoder.transform(customer_df['city_name_fa'])
        
        # Standardization of numerical features
        scaled_features = self.scaler.transform(customer_df[self.features_to_scale])
        
        # Convert the result to DataFrame
        customer_df[self.features_to_scale] = scaled_features

        return customer_df

    def _create_customer_features(self, df):
        df_temp = df.copy()
        df_temp['DateTime_CartFinalize'] = pd.to_datetime(df_temp['DateTime_CartFinalize'])
        
        # We consider the latest date as "today".
        self.current_date = df_temp['DateTime_CartFinalize'].max()

        # Group by customer and create new features
        customer_features = df_temp.groupby('ID_Customer').agg(
            order_count=('ID_Order', 'nunique'),
            unique_item_count=('ID_Item', 'nunique'),
            total_quantity=('Quantity_item', 'sum'),
            total_spent=('Amount_Gross_Order', 'sum'),
            max_order_value=('Amount_Gross_Order', 'max'),
            last_order_date=('DateTime_CartFinalize', 'max'),
            city_name_fa=('city_name_fa', 'first')
        ).reset_index()

        # Create Derived Features
        # RFM (Recency, Frequency, Monetary)
        
        customer_features['average_order_value'] = customer_features['total_spent'] / customer_features['order_count']
        customer_features['average_quantity_per_order'] = customer_features['total_quantity'] / customer_features['order_count']
        customer_features['item_variety_ratio'] = customer_features['unique_item_count'] / customer_features['total_quantity']
        
        # Replacing infinite values with 0 (without using inplace)
        customer_features['item_variety_ratio'] = customer_features['item_variety_ratio'].replace([np.inf, -np.inf], 0)

        # create time features (Recency)
        customer_features['days_since_last_order'] = (self.current_date - customer_features['last_order_date']).dt.days
        customer_features = customer_features.drop(columns=['last_order_date'])

        return customer_features
```

```python
from preprocessor import Preprocessor

# 2. Load Data
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

df = pd.read_csv('orders.csv')

# 3. sample for quick test
print("sample for quick test...")
df_sample = df.sample(n=50000, random_state=42)

# 4. preprocess data
print("preprocess data ...")
preprocessor = Preprocessor()
preprocessor.fit(df_sample)
processed_data = preprocessor.transform(df_sample)

print(f"The final data for clustering has {processed_data.shape[0]} customers and {processed_data.shape[1]} features.")
print("Final columns:")
print(processed_data.columns.tolist())

# 5. show info from processed data
print("\nSample of precessed data:")
print(processed_data.head())
print(f"\nStatistical Info:")
print(processed_data.describe())

# 6. Finding the optimal number of clusters (for quick test)
print("\nFinding the optimal number of clusters (this step may take a few minutes.)...")
X = processed_data.drop(columns=['ID_Customer'])

def find_optimal_clusters(data, max_k=6):  # for quick test
    silhouette_scores = []
    k_range = range(2, max_k + 1)
    
    for k in k_range:
        print(f"Processing for find K={k}...")
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10, verbose=0)
        labels = kmeans.fit_predict(data)
        score = silhouette_score(data, labels)
        silhouette_scores.append(score)
        print(f"K = {k}, Silhouette Score = {score:.4f}")
    
    # show plot
    plt.figure(figsize=(10, 6))
    plt.plot(k_range, silhouette_scores, 'bo-')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Score for Optimal K')
    plt.grid(True)
    plt.show()
    
    best_k = k_range[np.argmax(silhouette_scores)]
    print(f"bedt K: {best_k} with Silhouette Score: {max(silhouette_scores):.4f}")
    return best_k

optimal_k = find_optimal_clusters(X, max_k=6)

# 7.train final model with
print(f"\ntrain final model with K={optimal_k}...")
final_kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = final_kmeans.fit_predict(X)

# 8. final score
final_score = silhouette_score(X, clusters)
normalized_score = 100 * (final_score + 1) / 2

print(f"\n--- final result ---")
print(f"Silhouette Score: {final_score:.4f}")
print(f"Normalized Score: {normalized_score:.2f}")

if normalized_score >= 85:
    print("✅Your model has passed the score threshold.!")
else:
    print("❌Your model has not reached the required score. It needs improvement.")

# 9. save result
results_df = processed_data[['ID_Customer']].copy()
results_df['Cluster'] = clusters
results_df.to_csv('customer_clusters.csv', index=False)

# 10. Displaying the distribution of customers in clusters
cluster_counts = results_df['Cluster'].value_counts().sort_index()
print("\ndistribution of customers in clusters:")
for cluster, count in cluster_counts.items():
    print(f"cluster {cluster}: {count} customer ({count/len(results_df)*100:.1f}%)")

# 11. Analysis of Cluster Centers for a Better Understanding of Each Cluster
print("\nAnalysis of Cluster Centers:")
cluster_centers = final_kmeans.cluster_centers_
feature_names = X.columns

for i, center in enumerate(cluster_centers):
    print(f"\ncluster {i}:")
    for j, feature in enumerate(feature_names):
        print(f"  {feature}: {center[j]:.3f}")
```

###  Gaussian Mixture Model - GMM &

```python
# --- 1. SETUP: Import Libraries and Define X ---
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.mixture import GaussianMixture
from sklearn.neighbors import NearestNeighbors
from preprocessor import Preprocessor 
import matplotlib.pyplot as plt

# Load Data and Sample (for faster execution, as in your previous steps)
df = pd.read_csv('orders.csv')
# Sampling 50,000 rows for quick testing and to manage complexity of Hierarchical/DBSCAN
df_sample = df.sample(n=50000, random_state=42) 

# Preprocess Data
print("Starting preprocessing...")
preprocessor = Preprocessor()
preprocessor.fit(df_sample)
processed_data = preprocessor.transform(df_sample)

# Define X (Features ready for clustering)
# Removing the non-feature ID column
X = processed_data.drop(columns=['ID_Customer'])
# Using K=6, the optimal number found in K-Means analysis
OPTIMAL_K = 6 

print(f"Data X is ready for clustering. Shape: {X.shape}")

# --- 2. Gaussian Mixture Model (GMM) ---
print("\n" + "="*50)
print(f"GMM Clustering (K={OPTIMAL_K})")
print("="*50)

# GMM assumes clusters are generated from a mixture of Gaussian distributions.
# It uses the Expectation-Maximization (EM) algorithm.
try:
    gmm_final = GaussianMixture(n_components=OPTIMAL_K, 
                                random_state=42, 
                                n_init=10)
    gmm_final.fit(X)
    gmm_labels = gmm_final.predict(X)
    
    gmm_score = silhouette_score(X, gmm_labels)
    print(f"GMM Final Silhouette Score: {gmm_score:.4f}")
except Exception as e:
    print(f"Error during GMM: {e}")


# --- 3. Hierarchical Clustering (Agglomerative) ---
# Hierarchical clustering builds a hierarchy of clusters (dendrogram).
# Agglomerative (Bottom-Up) is used here, starting with individual points.
print("\n" + "="*50)
print(f"Hierarchical Clustering (Agglomerative) (K={OPTIMAL_K})")
print("="*50)

try:
    # 'ward' linkage minimizes the variance of the clusters being merged.
    hac_model = AgglomerativeClustering(n_clusters=OPTIMAL_K, 
                                        linkage='ward')
    hac_labels = hac_model.fit_predict(X)
    
    hac_score = silhouette_score(X, hac_labels)
    print(f"Hierarchical Silhouette Score: {hac_score:.4f}")
except Exception as e:
    print(f"Error during Hierarchical Clustering: {e}")


# --- 4. DBSCAN Clustering (Density-Based) ---
# DBSCAN finds high-density areas separated by low-density areas. 
# It does not require a fixed K and identifies outliers (labeled as -1).
print("\n" + "="*50)
print("DBSCAN Clustering (Density-Based)")
print("="*50)

try:
    # 1. Parameter Estimation
    # min_samples: minimum points required to form a dense region (core point). 
    # A common heuristic is 2 * number of dimensions.
    min_samples = 2 * X.shape[1] 
    
    # eps (Epsilon): maximum distance between two samples for one to be considered 
    # as in the neighborhood of the other. (0.5 is a common start for standardized data)
    estimated_eps = 0.5 

    dbscan_model = DBSCAN(eps=estimated_eps, min_samples=min_samples)
    dbscan_labels = dbscan_model.fit_predict(X)

    # Analysis of DBSCAN results
    # Cluster count excludes the noise cluster (-1)
    n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
    n_noise = list(dbscan_labels).count(-1)
    percent_noise = (n_noise/len(X)) * 100

    print(f"DBSCAN Parameters: eps={estimated_eps}, min_samples={min_samples}")
    print(f"Number of clusters found: {n_clusters}")
    print(f"Number of noise points (Outliers): {n_noise} ({percent_noise:.2f}%)")

    # Evaluate DBSCAN only on points that are assigned to a cluster (i.e., not noise)
    if n_clusters > 1:
        core_samples_mask = dbscan_labels != -1
        dbscan_score = silhouette_score(X[core_samples_mask], dbscan_labels[core_samples_mask])
        print(f"DBSCAN Silhouette Score (Excluding Noise): {dbscan_score:.4f}")
    else:
        print("DBSCAN performance is not optimal for these parameters (adjust eps/min_samples).")
        
except Exception as e:
    print(f"Error during DBSCAN: {e}")
```
